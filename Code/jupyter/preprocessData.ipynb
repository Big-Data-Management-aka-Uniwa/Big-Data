```python
# Libaries
import json
import pandas as pd
```

```python
# Datasets
PoliceKillingUS = None
PovertyUS = None
PopulationUS = None
USStates = None
FilteredPopulationUS = None
FilteredPoliceKillingUS = None
FilteredPovertyUS = None

PoliceKillingFinal2015 = None
PoliceKillingFinal2016 = None
PoliceKillingFinal = None
PoliceKillingFinalAvg = None
PovertyUSFinal = None
```

```python
"""
    Data Preprocessing 
"""
# [Step 1] Store the 3 datasets as DataFrame Objects 
PoliceKillingUS = pd.read_csv('../datasets/PoliceKillingsUS.csv', encoding='utf-8')
PoliceKillingUS.head()
```

```python
PovertyUS = pd.read_csv('../datasets/PovertyUS.csv', encoding='utf-8')
PovertyUS.head()
```

```python
PopulationUS = pd.read_csv('../datasets/PopulationUS_2015.csv', encoding='latin1', delimiter=';')
PopulationUS.head()
```

```python
"""
    Filter Dataset's Dates 
"""
def filter_by_date():
    # [Step 2.1] Convert date to standard format for pandas
    PoliceKillingUS['date'] = pd.to_datetime(PoliceKillingUS['date'], format='%d/%m/%y') 
     
    # [Step 2.2] Select the range of dates that we want (period 2015-2016)
    FilteredPoliceKillingUS = PoliceKillingUS[(PoliceKillingUS['date'] >= '2015-01-01') & (PoliceKillingUS['date'] <= '2016-12-31')]

    # [Step 2.3] Select the range of years (period 2015-2016)
    FilteredPovertyUS = PovertyUS[(PovertyUS['Year'] >=  2015) & (PovertyUS['Year'] <= 2016)]

    return FilteredPoliceKillingUS, FilteredPovertyUS
```

```python
# [Step 2] Filter the dates of PoliceKilling & PovertyUS datasets (Keep only 2015-2016 data) 
FilteredPoliceKillingUS, FilteredPovertyUS = filter_by_date()
FilteredPoliceKillingUS.head()
```

```python
FilteredPovertyUS.head()
```

```python
"""
    Change US states from full names to 2-letter names 
"""
def rename_states():
    # [Step 3.1] Open US states dataset and store it to a dict
    with open('../datasets/us_states.json', 'r') as f:
        USStates = json.load(f)

    # [Step 3.2] Replace the full name to 2-letter name for every US state
    for name in PovertyUS['Name']:
        if name in USStates:
            FilteredPovertyUS['Name'] = FilteredPovertyUS['Name'].replace(name, USStates[name])
    
    return FilteredPovertyUS
```

```python
# [Step 3] Change the name of US States from full name to 2-letter name (eg. California --> CA)
FilteredPovertyUS = rename_states()
FilteredPovertyUS = FilteredPovertyUS[FilteredPovertyUS['Name'] != 'US']
FilteredPovertyUS.head()
```

```python
"""
    Calculate each US state's population 
"""
def calculate_population():
    # [Step 4.1] Group by 'State' and sum 'Total_Population'
    FilteredPopulationUS = PopulationUS.groupby('State')['Total_Population'].sum()

    # [Step 4.2] Convert the Series to a DataFrame
    FilteredPopulationUS = FilteredPopulationUS.reset_index()

    return FilteredPopulationUS
```

```python
# [Step 4] Calculate each US state's population in 2015
FilteredPopulationUS = calculate_population()
FilteredPopulationUS.head()
```

```python
"""
    Save Processed Datasets to .csv  
"""
def save_to_csv():
    # [Step 5.1] Save the processed datasets to .csv files
    FilteredPoliceKillingUS.to_csv('../processed_datasets/ProcessedPoliceKillingUS.csv')
    FilteredPovertyUS.to_csv('../processed_datasets/ProcessedPovertyUS.csv')
    FilteredPopulationUS.to_csv('../processed_datasets/ProcessedPopulationUS_2015.csv')
```

```python
# [Step 5] Save the Processed datasets to different .csv files
save_to_csv()
```

```python
"""
    Reduce the data of kills 
"""
def reduce_data_of_kills(df):
    # [Step 6.1] Groups data for every state
    grouped = df.groupby('state')  

    # [Step 6.2] Finds most frequent row for every state using most_frequent function
    summary = grouped.agg({ 
        'id': 'count', #counts the kills for this state
        'date': lambda x: most_frequent(x.dt.year),
        'manner_of_death': most_frequent,
        'armed': most_frequent,
        'age': most_frequent,
        'gender': most_frequent,
        'race': most_frequent,
        'city': most_frequent,
        'signs_of_mental_illness': most_frequent,
        'threat_level': most_frequent,
        'flee': most_frequent,
        'body_camera': most_frequent
        })

    # [Step 6.3] Rename the 'id' column to 'count'
    summary = summary.rename(columns={'id': 'count'}) 
        
    # [Step 6.4] Reset index to make 'state' a column
    summary = summary.reset_index()

    # [Step 6.5] Calculate total kills
    total_kills = df.shape[0] # get x dimension
    summary['percentage'] = (summary['count'] / total_kills) * 100
        
    return summary
    
# [Step 6.2.1] Finds the most frequent value
@staticmethod
def most_frequent(series):
    return series.mode().iloc[0]
```

```python
# [Step 6] Keep only the data of Police killing in 2015 and in 2016 
PoliceKillingFinal2015 = reduce_data_of_kills(pd.DataFrame(FilteredPoliceKillingUS[(FilteredPoliceKillingUS['date'] >= '2015-01-01') & (FilteredPoliceKillingUS['date'] <= '2015-12-31')]))
PoliceKillingFinal2015.head()
```

```python
PoliceKillingFinal2016 = reduce_data_of_kills(pd.DataFrame(FilteredPoliceKillingUS[(FilteredPoliceKillingUS['date'] >= '2016-01-01') & (FilteredPoliceKillingUS['date'] <= '2016-12-31')]))
PoliceKillingFinal2016.head()
```

```python
# [Step 7] Merge the 2 Police Killing datasets 
PoliceKillingFinal = pd.concat([PoliceKillingFinal2015, PoliceKillingFinal2016], ignore_index=True)
PoliceKillingFinal.head()
```

```python
PoliceKillingFinal.tail()
```

```python
"""
    Calculate the Poverty Average by state
"""
def poverty_averages():
    # [Step 8.1] Filter data for the years 2015 and 2016
    poverty_2015 = FilteredPovertyUS[FilteredPovertyUS['Year'] == 2015]
    poverty_2016 = FilteredPovertyUS[FilteredPovertyUS['Year'] == 2016]
        
    # [Step 8.2] Merge the two datasets on 'ID'
    merged_data = pd.merge(poverty_2015, poverty_2016, on='ID', suffixes=('_2015', '_2016'))
       
    # [Step 8.3] Calculate average values for the required columns
    merged_data['Poverty Universe Avg'] = (merged_data['Poverty Universe_2015'] + merged_data['Poverty Universe_2016']) / 2
    merged_data['Number in Poverty Avg'] = (merged_data['Number in Poverty_2015'] + merged_data['Number in Poverty_2016']) / 2
    merged_data['Percent in Poverty Avg'] = (merged_data['Percent in Poverty_2015'] + merged_data['Percent in Poverty_2016']) / 2

    # [Step 8.4] Select and rename the columns
    PovertyUSFinal = merged_data[['ID', 'Name_2015', 'Poverty Universe Avg', 'Number in Poverty Avg', 'Percent in Poverty Avg']]
    PovertyUSFinal.rename(columns={'Name_2015': 'Name'}, inplace=True)

    # [Step 8.5] Save the final poverty dataset to a CSV file
    PovertyUSFinal.to_csv('../datasets/PovertyUS_Avg_2015_2016.csv', index=False)

    return PovertyUSFinal
```

```python
# [Step 8] Calculate the Poverty Average by state (PovertyUS_Avg_2015_2016.csv) 
PovertyUSFinal = poverty_averages()
PovertyUSFinal.head()
```

```python
"""
        Calculate the Killings Average by state
"""
def killing_averages():
    # [Step 9.1] Merge the datasets for 2015 and 2016 police killings
    merged_data = pd.merge(PoliceKillingFinal2015, PoliceKillingFinal2016, on='state', suffixes=('_2015', '_2016'), how='outer')
    merged_data.head()

    # [Step 9.2] Calculate average values for the required columns
    merged_data['Avg Deaths'] = (merged_data['count_2015'].fillna(0) + merged_data['count_2016'].fillna(0)) / 2
    merged_data['Avg Deaths in percentage'] = (merged_data['percentage_2015'].fillna(0) + merged_data['percentage_2016'].fillna(0)) / 2

    # [Step 9.3] Select the final columns
    PoliceKillingFinalAvg = merged_data[['state', 'Avg Deaths', 'Avg Deaths in percentage']]

    # [Step 9.4] Exclude data for the entire US
    PoliceKillingFinalAvg = PoliceKillingFinalAvg[PoliceKillingFinalAvg['state'] != 'US']

    # [Step 9.5] Save the final killings dataset to a CSV file
    PoliceKillingFinalAvg.to_csv('../datasets/PolliceKillingUS_Avg_2015_2016.csv', index=False)

    return PoliceKillingFinalAvg
```

```python
# [Step 9] Calculate the Killings Average by state (PoliceKillingUs_Avg_2015_2016.csv)  
PoliceKillingFinalAvg = killing_averages()
PoliceKillingFinalAvg.head()
```
